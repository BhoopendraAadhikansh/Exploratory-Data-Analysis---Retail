##**Data Analysis by Region wise**

```
sale_region=mrket_data[['Region','Sales']].groupby('Region').sum().unstack(fill_value=0)
sale_region.plot(kind='pie',autopct='%1.1f%%',startangle=-40,shadow=True,explode=(0.02,0.02,0.02,0.02))
plt.title('Sales per Regions',fontsize=25,weight='bold')
plt.tight_layout()
plt.show()
```
